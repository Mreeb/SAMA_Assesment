```python
train_data, test_data = train_test_split(ratings_data, test_size=0.2, random_state=42)

reader = Reader(line_format='user item rating', sep='\t')
data = Dataset.load_from_df(train_data.unstack().reset_index(), reader)

sim_options = {
    'name': 'cosine',
    'user_based': False  # Item-based collaborative filtering
}
model = KNNBasic(sim_options=sim_options)

model.fit(data.build_full_trainset())

test_set = test_data.unstack().reset_index().values.tolist()
predictions = model.test(test_set)
print("Item-based CF Model - RMSE on test set:", accuracy.rmse(predictions))
```

```python
from surprise import SVD, KNNBasic, KNNWithMeans
from surprise.model_selection import cross_validate

algorithms = [
    SVD(),
    KNNBasic(sim_options={'user_based': False}),
    KNNWithMeans(sim_options={'user_based': False})
]

# Perform cross-validation for each algorithm
results = {}
for algorithm in algorithms:
    print(f"Evaluating {algorithm.__class__.__name__}...")
    cv_results = cross_validate(algorithm, data, measures=['RMSE', 'MAE'], cv=5, verbose=True)
    results[algorithm.__class__.__name__] = cv_results

for algorithm, result in results.items():
    print(f"{algorithm}:")
    print("  RMSE:", round(result['test_rmse'].mean(), 4))
    print("  MAE:", round(result['test_mae'].mean(), 4))
```

```python
def get_recommendations_for_user(user_id, model, movie_titles, num_recommendations=5):
    user_ratings = train_data.loc[user_id]
    unrated_movies = user_ratings[user_ratings == 0].index.tolist()

    predictions = []
    for movie_id in unrated_movies:
        prediction = model.predict(str(user_id), str(movie_id))
        predictions.append((movie_id, prediction.est))

    predictions.sort(key=lambda x: x[1], reverse=True)

    top_recommendations = predictions[:num_recommendations]

    recommendations = [(movie_titles[int(movie_id)], rating) for movie_id, rating in top_recommendations]

    return recommendations

user_id = 200
recommendations = get_recommendations_for_user(user_id, model, movie_titles)
print("Top Recommendations for User", user_id, ":")
for i, (movie_title, rating) in enumerate(recommendations):
    print(f"{i+1}. {movie_title} (Predicted Rating: {rating:.2f})")
```

# USing Cosine Simmilarity

```python
from sklearn.metrics.pairwise import cosine_similarity

ratings_data.fillna(ratings_data.mean(), inplace=True)

normalized_ratings = ratings_data.apply(lambda x: (x - x.mean()) / (x.max() - x.min()), axis=1)

similarity_matrix = cosine_similarity(normalized_ratings, normalized_ratings)

def recommend_next_movie(input_movie):
    movie_index = movie_titles[movie_titles == input_movie].index[0]
    similarity_scores = similarity_matrix[movie_index]
    similar_users_indices = similarity_scores.argsort()[::-1][1:]  # Exclude the input movie itself
    predicted_ratings = ratings_data.iloc[similar_users_indices].mean()
    predicted_ratings = predicted_ratings[ratings_data.iloc[movie_index] == 0]
    print(predicted_ratings)
    top_recommendations = predicted_ratings.sort_values(ascending=False).index
    return top_recommendations[0]  # Return the top recommended movie
```

```python
# Example usage
input_movie = str(input("Enter Movie Name: "))
print("Entered Movie: ", input_movie)
recommended_movie_index = recommend_next_movie(input_movie)
recommended_movie_title = movie_titles.iloc[recommended_movie_index]
print("Recommended movie:", recommended_movie_title)
```

# Lets Try out K NearestNeighbors

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.neighbors import NearestNeighbors
ratings_data = movies_df.drop('title', axis=1)

similarity_matrix = cosine_similarity(ratings_data)

knn_model = NearestNeighbors(metric='cosine', algorithm='brute')
knn_model.fit(ratings_data)


def recommend_next_movie(input_movie):
    input_index = movies_df[movies_df['title'] == input_movie].index[0]
    _, indices = knn_model.kneighbors(ratings_data.iloc[input_index].values.reshape(1, -1))

    recommended_index = indices[0][1]
    recommended_movie = movies_df.iloc[recommended_index]['title']

    return recommended_movie

input_movie = "Austin Powers: The Spy Who Shagged Me (1999)"
recommended_movie = recommend_next_movie(input_movie)
print("Recommended movie:", recommended_movie)
```
